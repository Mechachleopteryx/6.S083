```julia
function run_cos(x0, N)
    x = zeros(N + 1)
    x[1] = x0
    
    for n = 1:N
        x[n+1] = cos(x[n])
    end
    return x
end
```

```julia
data = run_cos(0.3, 100)
```

```julia
plot(data, m=:o)   # marker, :x is another possibilty
```

```julia
plot(data, m=:square)   # marker, :x is another possibilty
```

```julia
plot(data[10:end], m=:o)   # marker, :x is another possibilty
```

```julia
fixed_point = data[end]
```

Distance of each data point from this fixed_point:

```julia
data - fixed_point
```

```julia
new_data = data .- fixed_point  # broadcasting: point by point / element by element
```

```julia
plot(new_data, m=:o)
```

```julia
new_data2 = abs(new_data)
```

```julia
new_data2 = abs.(new_data)
```

```julia
plot(new_data2, m=:o)
```

Maybe $1/n$ behaviour:

```julia
plot(1 ./ new_data2, m=:o)
```

```julia
plot(new_data2 .+ 0.00001, yscale=:log, m=:o)
```

```julia
function dice_experiment(sides, N)
    data = rand(1:sides, N)

    counts = zeros(Int, sides)

    for result in data 
        counts[result] += 1
    end
    
    return counts
end
```

```julia
counts = dice_experiment(6, 100)
```

```julia
using Plots
```

```julia
counts = dice_experiment(6, 100)
bar(counts)
```
